```python
#these are all the imports you need
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
from scipy.stats import t
```

# Priors and Likelihoods

Standard normal with mean 0 and variance 1

```python
x = np.linspace(-10, 45, 1000)
y_1 = norm.pdf(x)
plt.xlabel("The support(?)")
plt.ylabel("Probability")
plt.plot(x,y_1/max(y_1))
```

Standard normal with mean 10 and variance 1

```python
y_2 = norm.pdf(x, 10, 1)
plt.xlabel("The support(?)")
plt.ylabel("Probability")
plt.plot(x,y_2/max(y_2))
```

Student's with 2 degrees of freedom centered on 0


```python
y_3 = t.pdf(x, 2)
plt.xlabel("The support(?)")
plt.ylabel("Probability")
plt.plot(x,y_3/max(y_3))
```

Student's with 2 degrees of freedom centered on 10

```python
y_4 = t.pdf(x, 2, 10)
plt.xlabel("The support(?)")
plt.ylabel("Probability")
plt.plot(x,y_4/max(y_4))
```

# Posteriors

Prior: 1 \\
Likelihood: 2

```python
posterior = y_1*y_2
plt.plot(x, y_1/max(y_1), color = "black", label = "Prior", ls = '--')
plt.plot(x, y_2/max(y_2), color = "black", label = "Likelihood", ls = '--')
plt.plot(x, posterior/max(posterior), color = 'blue', label = "Posterior")
plt.legend()
plt.xlabel("The support(?)")
plt.ylabel("Probability")
plt.show()
```

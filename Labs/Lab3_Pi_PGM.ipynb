```python
import numpy as np
import pylab as pl
#!pip install astropy
import astropy
from astropy import units as u


%pylab inline
```

```python
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
```

```python
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
x = np.random.rand(1000)*2 -1
y = np.random.rand(1000)*2 -1
ax.plot(x,y, 'o', alpha = 0.2)
```

```python
def isin(x,y,l):
  if np.sqrt(x**2 + y**2) <= l:
    return True
  return False
```

```python
l_everywhere = 1
np.random.seed(320)
counter = 0
N = 1000
for i in range(N):
  x, y = (np.random.rand() * 2 - 1), np.random.rand()*2 - 1
  if isin(x, y, l_everywhere):
    counter +=1
counter
```

```python
pseudoPi = counter/N*4
pseudoPi
```
